```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U(x):
    return (x**2)*(x-1)*(x+1)
```

```python
def dU(x):
    return 4*(x**3)-2*x
```

```python
def gradiente(dU, x0, alpha=0.1, eps=0.01, max=1000):
    x=x0
    lista = [x]
    for _ in range(max):
        x1=x-alpha*dU(x)
        lista.append(x1)
        if abs(x1-x)<eps:
            break
        x=x1
    return lista
```

```python
#GrÃ¡fico potencial
eixo_x=np.linspace(-2.4, 2.4, 200)
eixo_y=U(eixo_x)
```

```python
#ParÃ¢metros do cÃ³digo
x0=2
alpha=0.1
eps=0.01
maximo=1000

#Calculando o gradiente 
caminho=gradiente(dU, x0, alpha, eps, maximo)

#Primeira CondiÃ§Ã£o
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = xÂ²(x-1)(x+1)')  # Plota a funÃ§Ã£o
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='TrajetÃ³ria da partÃ­cula')  # TrajetÃ³ria
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-1,14)
plt.title('Gradiente Descendente para U(x) = xÂ²(x-1)(x+1)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#ParÃ¢metros do cÃ³digo
x0=2
alpha=0.01
eps=0.01
maximo=1000

#Calculando o gradiente 
caminho=gradiente(dU, x0, alpha, eps, maximo)

#Segunda CondiÃ§Ã£o
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = xÂ²(x-1)(x+1)')  # Plota a funÃ§Ã£o
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='TrajetÃ³ria da partÃ­cula')  # TrajetÃ³ria
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-1,14)
plt.title('Gradiente Descendente para U(x) = xÂ²(x-1)(x+1)')
plt.legend()
plt.grid(True)
plt.show()
```
