```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U(x):
    return x**2 - 1
```

```python
def dU(x):
    return 2*x
```

```python
def gradiente(dU, x0, alpha=0.1, eps=0.01, max=1000):
    x=x0
    lista = [x]
    for _ in range(max):
        x1=x-alpha*dU(x)
        lista.append(x1)
        if abs(x1-x)<eps:
            break
        x=x1
    return lista
```

```python
#Calculando o gradiente 
caminho=gradiente(dU, x0)
```
