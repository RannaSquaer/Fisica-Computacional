# Exercício 3

Repita o exercício 2, mas agora vamos manipular a altura dos mínimos somando uma reta em $U(x)$, tal que a função agora é $U(x)=x^2(x−1)(x+1)+(x/4)$. O que acontece? O que você pode concluir sobre a escolha da taxa de aprendizado $\alpha$?

---

Precisamos definir a nossa função $U(x)$ e sua derivada $dU(x)$:

$$ U(x)=x^2(x−1)(x+1) + \frac{x}{4} $$

$$ dU(x) = 4x^3 - 2x + \frac{1}{4}$$

Nesse caso, os mínimos não estão simétricos como no caso anterior porque a função foi inclinada com a adição de $\frac{x}{4}$, uma reta.

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos a função potencial U(x), que queremos minimizar
def U(x):
    return (x**2)*(x-1)*(x+1)+(x/4)
```

```python
# Derivada de U(x), que representa o gradiente da função
def dU(x):
    return 4*(x**3)-2*x + (1/4)
```

```python
# Função que implementa o algoritmo de gradiente descendente
# Esse algoritmo simula uma partícula descendo pela "curva" da função U(x), parando no mínimo local
def gradiente(dU, x0, alpha=0.1, eps=0.01, max=1000):
    x=x0 # posição inicial da partícula
    lista = [x] # armazena as posições da partícula
    for _ in range(max):
        x1=x-alpha*dU(x) # novo ponto
        lista.append(x1) # adiciona o novo ponto à lista de posições
        if abs(x1-x)<eps: # critério de parada baseado na variação
            break
        x=x1 # atualiza x para a próxima iteração
    return lista # retorna a lista de posições da partícula que define a sua trajetória
```

```python
# Geração de pontos para desenhar a curva da função U(x)
eixo_x=np.linspace(-2.4, 2.4, 200)
eixo_y=U(eixo_x)
```

```python
#===========================
# Primeira simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=2
alpha=0.1
eps=0.001
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x)
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x²(x-1)(x+1)+(x/4)')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-1,13)
plt.title('Gradiente Descendente para U(x) = x²(x-1)(x+1)+(x/4) (Primeira simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

A taxa de aprendizado foi definida como $\alpha = 0.1$, o algoritmo rapidamente converge para o mínimo localizado à esquerda de $x = 0$, mesmo partindo de $x_0 = 2$. Isso indica que o valor de $\alpha$ foi grande o suficiente para a partícula ultrapassar a barreira em $x = 0$. Este resultado é similar ao encontrado no Exercício 2, para os mesmos parâmentros, mesmo com a assimetria dos mínimos.

```python
#===========================
# Segunda simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=2
alpha=0.01
eps=0.01
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x)
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x²(x-1)(x+1)+(x/4)')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-1,13)
plt.title('Gradiente Descendente para U(x) = x²(x-1)(x+1)+(x/4) (Segunda simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Neste caso, com a taxa de aprendizado menor $\alpha=0.01$ resultou em uma descida mais suave. O resultado é similar ao encontrado no Exercício 2.

```python
#===========================
# Terceira simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=2
alpha=0.1399
eps=0.01
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x)
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x²(x-1)(x+1)+(x/4)')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-1,13)
plt.title('Gradiente Descendente para U(x) = x²(x-1)(x+1)+(x/4) (Terceira simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Neste caso, com a taxa de aprendizado $\alpha=0.1399$ a partícula tem um comportamento oscilatório, no entanto, após algumas interações atinge o mínimo do lado esquerdo de $x=0$. No Exercício 2, ela atingiu o mínimo da direita.

```python
#===========================
# Quarta simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=1
alpha=0.6
eps=0.001
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x)
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x²(x-1)(x+1)+(x/4)')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-0.5,0.28)
plt.title('Gradiente Descendente para U(x) = x²(x-1)(x+1)+(x/4) (Quarta simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Quando utilizamos um valor alto para a taxa de aprendizagem $\alpha=0.6$, é possível notar que o comportamento oscilatório se intensifica, e a partícula não consegue estabilidade e precisão.

```python
#===========================
# Quinta simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=1
alpha=0.03
eps=0.001
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x)
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x²(x-1)(x+1)+(x/4)')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-0.1,0.28)
plt.title('Gradiente Descendente para U(x) = x²(x-1)(x+1)+(x/4) (Quinta Simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Para uma taxa de aprendizagem de $\alpha=0.03$, é possível notar que a tragetória segue a curvatura da função e apresenta uma boa precisão.

# Análise final 
---

Essa simulação mostra que um aumento moderado da taxa de aprendizado pode acelerar a convergência, desde que a estabilidade ainda seja mantida.

- A função $U(x)=x^2(x−1)(x+1)+(x/4)$ tem dois mínimos locais assimétricos, sendo o da esquerda mais profundo.
- Com $\alpha = 0.03$ e $x_0 = 1$, a partícula converge para o mínimo da direita de forma lenta e estável.
- Com $\alpha = 0.6$ e $x_0 = 1$, a partícula não se estabiliza e continua oscilando ao redor do máximo local.
- A posição inicial influencia fortemente qual mínimo será alcançado.
- A escolha adequada da taxa de aprendizado $\alpha$ é essencial para evitar oscilações e garantir a convergência eficiente.
