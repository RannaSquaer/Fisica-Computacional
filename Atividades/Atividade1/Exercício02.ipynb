# Exercício 2

Repita o exercício 1 para a função $U(x)=x^2(x−1)(x+1)$. Esta função tem dois mínimos globais. Use $x_0=2$ e tente ajustar $\alpha$ para tentar fazer o código convergir ora num mínimo, ora no outro. O que acontece? O que você pode concluir sobre a escolha da taxa de aprendizado $\alpha$?

---


Precisamos definir a nossa função $U(x)$ e sua derivada $dU(x)$:

$$ U(x)=x^2(x−1)(x+1) $$

$$ dU(x) = 4x^3 - 2x $$

Para determinar as raízes temos que $dU(x) = 0$, ou seja, temos a solução trivial quando $x=0$, e temos $x = \pm \frac{\sqrt{2}}{2}$.

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos a função potencial U(x), que queremos minimizar
def U(x):
    return (x**2)*(x-1)*(x+1)
```

```python
# Derivada de U(x), que representa o gradiente da função 
def dU(x):
    return 4*(x**3)-2*x
```

```python
# Função que implementa o algoritmo de gradiente descendente
# Esse algoritmo simula uma partícula descendo pela "curva" da função U(x), parando no mínimo local
def gradiente(dU, x0, alpha=0.1, eps=0.01, max=1000):
    x=x0 # posição inicial da partícula
    lista = [x] # armazena as posições da partícula
    for _ in range(max):
        x1=x-alpha*dU(x) # novo ponto
        lista.append(x1) # adiciona o novo ponto à lista de posições
        if abs(x1-x)<eps: # critério de parada baseado na variação
            break
        x=x1 # atualiza x para a próxima iteração
    return lista # retorna a lista de posições da partícula que define a sua trajetória
```

```python
# Geração de pontos para desenhar a curva da função U(x)
eixo_x=np.linspace(-2.4, 2.4, 200)
eixo_y=U(eixo_x)
```

```python
#===========================
# Primeira simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=2
alpha=0.1
eps=0.01
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x) 
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x²(x-1)(x+1)')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-1,14)
plt.title('Gradiente Descendente para U(x) = x²(x-1)(x+1) (Primeira simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

No caso em que definimos a taxa de aprendizado em $\alpha = 0.1$, o algoritmo rapidamente converge para o mínimo localizado à esquerda de $x = 0$, mesmo partindo de $x_0 = 2$. Isso indica que o valor de $\alpha$ foi grande o suficiente para a partícula ultrapassar a barreira em $x = 0$.

```python
#===========================
# Segunda simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=2
alpha=0.01
eps=0.01
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x) 
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x²(x-1)(x+1)')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-1,14)
plt.title('Gradiente Descendente para U(x) = x²(x-1)(x+1) (Segunda simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Neste caso, com a taxa de aprendizado menor $\alpha=0.01$ resultou em uma descida mais suave. A partícula permaneceu na região direita de $x = 0$ e convergiu para o mínimo da direita, sem energia suficiente para ultrapassar $x = 0$.

```python
#===========================
# Terceira simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=2
alpha=0.1399
eps=0.01
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x) 
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x²(x-1)(x+1)')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.ylim(-1,14)
plt.title('Gradiente Descendente para U(x) = x²(x-1)(x+1) (Terceira simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Neste caso, com a taxa de aprendizado $\alpha=0.1399$ a partícula apresenta um comportamento oscilatório, porém depois de algumas iterações ela fica restrita à direita de $x=0$.

# Análise final
---

Estamos estudando a função $U(x) = x^2(x−1)(x+1)$, que possui dois mínimos simétricos em $x= \pm \frac{\sqrt{2}}{2}$ e um ponto de máximo em $x = 0$. A simulação foi realizada com o ponto inicial $x_0 = 2$ e três taxas de aprendizado diferentes: $\alpha = 0.1$, $\alpha = 0.01$ e $\alpha=0.1399$.

- A taxa de aprendizado $\alpha$ influencia tanto a rapidez quanto a estabilidade da convergência.
- $\alpha$ pequeno (como 0.01) garantem suavidade, $\alpha$ moderado (como 0.1) aceleram a descida, e $\alpha$ grande (como 0.1399) causam oscilações.
