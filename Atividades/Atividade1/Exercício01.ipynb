# Exercício 1

Implemente o algoritmo de gradiente descendente para encontrar o mínimo da função $U(x)=x^2−1$. É um caso bem simples para o qual sabemos a solução exata. Ilustre o algoritimo com um gráfico mostrando a função $U(x)$ e a trajetória da partícula. Use inicialmente uma taxa de aprendizado $\alpha=0.1$ e uma tolerância $\epsilon=0.01$. O número máximo de iterações deve ser 1000. A posição inicial da partícula deve ser $x_0=5$. Depois, varie estes parâmetros para ver como eles afetam a convergência do algoritmo.

---

Precisamos definir a nossa função $U(x)$ e sua derivada $dU(x)$:

$$U(x) = x^2 - 1$$

$$dU(x) = 2x$$

Para determinar as raízes temos $dU(x) = 0$, ou seja, nesse caso em espefícico, quando $x=0$.

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos a função potencial U(x), que queremos minimizar
def U(x):
    return x**2 - 1
```

```python
# Derivada de U(x), que representa o gradiente da função 
def dU(x):
    return 2*x
```

```python
# Função que implementa o algoritmo de gradiente descendente
# Esse algoritmo simula uma partícula descendo pela "curva" da função U(x), parando no mínimo local

def gradiente(dU, x0, alpha=0.1, eps=0.01, max=1000):
    x=x0 # posição inicial da partícula
    lista = [x] # armazena as posições da partícula
    for _ in range(max):
        x1=x-alpha*dU(x) # novo ponto
        lista.append(x1) # adiciona o novo ponto à lista de posições
        if abs(x1-x)<eps: # critério de parada baseado na variação
            break
        x=x1 # atualiza x para a próxima iteração
    return lista # retorna a lista de posições da partícula que define a sua trajetória
```

```python
# Geração de pontos para desenhar a curva da função U(x)
eixo_x=np.linspace(-8,8,200)
eixo_y=U(eixo_x)
```

```python
#===========================
# Primeira simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=5
alpha=0.1
eps=0.01
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x) 
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x² - 1')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.title('Gradiente Descendente para U(x) = x² - 1 (Primeira simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Nesta primeira simulação, a partícula parte de $x_0=5$ com $\alpha=0.1$, $\epsilon=0.01$ e máximo de iterações sendo 1000 . O algoritmo está convergindo corretamente para o mínimo da função, que é $𝑥=0$, com a partícula fazendo passos cada vez menores até parar.

```python
# Segunda simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=7
alpha=0.3
eps=0.01
maximo=1000

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x) 
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x² - 1')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.title('Gradiente Descendente para U(x) = x² - 1 (Segunda simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Na segunda simulação, , a partícula parte de $x_0=7$ com $\alpha=0.3$, $\epsilon=0.01$ e máximo de iterações sendo 1000. O algoritmo também está convergindo para o mínimo da função, porém, com "saltos" maiores, o que tira a precisão da tragetória da partícula em relação a curva esperada da função. 

```python
#===========================
# Terceira simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=7
alpha=0.2
eps=0.001
maximo=200

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x)  
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x² - 1')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.title('Gradiente Descendente para U(x) = x² - 1 (Terceira simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Nessa terceira simulação, a partícula parte de $x_0=7$ com $\alpha=0.2$, $\epsilon=0.001$ e máximo de iterações sendo 200. A trajetória é mais estável que a anterior, e atinge o mínimo com precisão maior.

```python
#===========================
# Quarta simulação
#===========================
# Parâmetros: posição inicial, taxa de aprendizado, tolerância e máximo de iterações
x0=8
alpha=0.5
eps=0.001
maximo=200

# Simula o movimento da "partícula" que está descendo a curva da função potencial U(x)
caminho=gradiente(dU, x0, alpha, eps, maximo)

# Plot da função e trajetória da partícula
plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, label='U(x) = x² - 1')  # Plota a função
plt.plot(caminho, [U(x) for x in caminho], 'ro--', label='Trajetória da partícula')  # Trajetória
plt.xlabel('x')
plt.ylabel('U(x)')
plt.title('Gradiente Descendente para U(x) = x² - 1 (Quarta simulação)')
plt.legend()
plt.grid(True)
plt.show()
```

Na quinta simulação, a partícula parte de $x_0=7$ com $\alpha=0.01$, $\epsilon=0.001$ e máximo de iterações sendo 200. A partícula se move lentamente em direção ao mínimo, apresentando uma maior precisão, pois a tragetória acontece de forma suave e acompanha a curvatura da função.

# Análise final
---

Observando os resultados gráficos das diferentes condições, podemos tirar algumas conclusões:

Em todas as simulações, a trajetória converge para o ponto de mínimo da função $U(x)$, que é $x = 0$.
- A velocidade de convergência depende fortemente da taxa de aprendizado ($\alpha$).
- Taxas muito baixas (como $\alpha = 0.01$) fazem o processo ser muito lento.
- Taxas muito altas (como $\alpha = 0.5$) podem fazer o pocesso ser mais rápido, porém a precisão é prejudicada.
- O número máximo de iterações é importante para evitar loops infinitos caso a convergência não ocorra.

Essa análise mostra como o ajuste dos parâmetros é crucial para o desempenho do método do gradiente descendente.
